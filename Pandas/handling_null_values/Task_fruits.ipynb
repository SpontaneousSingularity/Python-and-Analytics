```python
#new_df = df.fillna(-1,inplace= False)
new_df = df
```

```python
new_df.fillna(
    {
        "apple(1kg)": df["apple(1kg)"].mean(),
        "banana(1 dozen)": df["banana(1 dozen)"].mean(),
        "grapes(1kg)": df["apple(1kg)"].median(),
        "mango(1kg)": df["apple(1kg)"].median(),
        "Water Melons(1)": "Not Available"
    }
)
new_df
```

```python
new_df.ffill(inplace= False)
```

```python
new_df.dropna(thresh=4,inplace= True)
```

```python
new_df.to_csv("final_data.csv")
```
