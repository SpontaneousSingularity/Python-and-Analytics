```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style 
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = {
    # 12.5k units of smartTV sold in Q1 2024
    "Quarter":["Q1 2024","Q2 2024","Q3 2024","Q4 2024","Q1 2025"],
    "Smart TV":[12.5,10,13.8,15,6],
    "Fridge":[18,21,14,25,3],
    "Washing Machine":[10,15,7,14,7]
}

df_sales = pd.DataFrame(data)
df_sales
```

```python
plt.figure(figsize=(12,5))
plt.plot(df_sales["Quarter"],df_sales["Smart TV"], label = "Smart TV", color="purple")
plt.plot(df_sales["Quarter"],df_sales["Fridge"], label = "Fridge")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"], label = "Washing Machine")

plt.title("Product Sales from 2024 to 2025")
plt.xlabel("Quarter")
plt.ylabel("Units Sold in K")

plt.legend()
plt.show()
```

## Pie chart 

```python
total_units_sold = df_sales[["Smart TV", "Fridge", "Washing Machine", ]].sum()
total_units_sold
```

```python
total_units_sold.index
```

```python
plt.pie(total_units_sold, labels=total_units_sold.index, autopct="%1.1f%%", explode=(0.1, 0, 0), shadow=True,)

plt.title("Market Share")
plt.show()
```

## Bar Chart

```python
df_sales.plot(kind= "bar",x="Quarter")

#Get or set the current tick locations and labels of the x-axis.
#xticks rotation rotates x labels
plt.xticks(rotation = 0)
plt.show()
```

```python
df_sales.plot(kind= "barh",x="Quarter")
plt.xticks(rotation = 0)
plt.show()
```

```python
df_sales.plot(kind= "barh",x="Quarter",subplots = True)
plt.xticks(rotation = 0)
plt.show()
```

bar chart takes x axis as index in default
