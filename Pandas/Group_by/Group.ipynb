### Group Based on Custom function
<li>Days when Temperature was between 24 and 30</li>
<li>Days when it was between 16 and 23</li>
<li>Days with anything else</li>


```python
def group_fn(df,indx,col):
    if df [col].loc[indx] >=24 and df[col].loc[indx] <=30:
        return "24 - 30"
    elif df [col].loc[indx] >=14 and df[col].loc[indx] <=23:
        return "14 - 23"
    else:
        return "Others"
    
```

```python
group_new = df. groupby(lambda row : group_fn(df, row, "temperature"))
```

```python
for key, data in group_new:
    print(f" key: {key}\n{data}\n")
```
