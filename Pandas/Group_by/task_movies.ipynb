```python
def group_fn(df,indx,col):
    if df [col].loc[indx] >=1 and df[col].loc[indx] <=3.9:
        return "Poor"
    elif df [col].loc[indx] >=4 and df[col].loc[indx] <=7.9:
        return "Average"
    elif df [col].loc[indx] >=8 and df[col].loc[indx] <=10:
        return "Good"
    else:
        return "Others"
```

```python
imdb_group = df.groupby( lambda row: group_fn(df,row,"imdb_rating"))
```
